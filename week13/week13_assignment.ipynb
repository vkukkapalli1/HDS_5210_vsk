### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
epl.describe()
```

```python
#these histograms are created to show the distribution of home team shots(HS) and away team shots(AS)
#since most of my columns are int or float type, I chose the above two columns
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(epl['HS.x'], bins=15, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Home Team Shots', fontsize=16)
plt.xlabel('Home Team Shots', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(1, 2, 2)
plt.hist(epl['AS.x'], bins=15, color='orange', edgecolor='black', alpha=0.7)
plt.title('Histogram of Away Team Shots', fontsize=16)
plt.xlabel('Away Team Shots', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
teams = epl['teamId'].unique()
plt.figure(figsize=(10,6))

for i, team in enumerate(teams):
    team_data = epl[epl['teamId'] == team]
    plt.plot(team_data['round'], team_data['tot_points'], label=f'{team}', linewidth=2)

plt.title('Team Total Points Progression Over Time', fontsize=16)
plt.xlabel('Matches', fontsize=12)
plt.ylabel('Total Points', fontsize=12)
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

#i have created a line chart showing how teams progressed over different match rounds
```
